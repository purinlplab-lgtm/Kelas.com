```python
import pandas as pd
```

```python
# mengimmpor data
df = pd.read_csv('train_house.csv')

# Lihat 5 baris pertama untuk cek data
df.head()
```

```python
# Lihat jumlah baris sebelum menghapus NaN
print("Jumlah baris sebelum drop NaN:", len(df))

# Hapus baris yang mengandung NaN
df_clean = df.dropna()

# Lihat jumlah baris setelah drop NaN
print("Jumlah baris setelah drop NaN:", len(df_clean))

# Tampilkan 5 baris pertama dataset yang sudah bersih
df_clean.head()

# boleh disesuaikan handlingnya misal numerik kolom bsa pakai median di masing2 kolom numerik dan kategorikal kolom bisa pakai modus di masing2 kolom kategorik
```

```python
# Pisahkan kolom numerikal
df_numerical = df.select_dtypes(include=['int64', 'float64'])

# Pisahkan kolom kategorikal
df_categorical = df.select_dtypes(include=['object', 'category'])

# Tampilkan info singkat
print("Kolom Numerikal:")
print(df_numerical.columns)

print("\nKolom Kategorikal:")
print(df_categorical.columns)
```

```python
#select_dtypes(include=[...]) → memilih kolom berdasarkan tipe datanya.
#•	'int64', 'float64' → numerikal
#•	'object', 'category' → kategorikal
#Hasilnya:
#•	df_numerical → hanya berisi kolom angka, siap untuk histogram, boxplot, scatter plot, dll.
#•	df_categorical → hanya berisi kolom kategori, siap untuk bar plot, pie chart, dll.
```

```python
import matplotlib.pyplot as plt
```

```python
categorical_cols = ['Utilities', 'HouseStyle', 'ExterQual']
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    df[col].value_counts().plot(kind='bar', color='#1F4E79')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
import seaborn as sns
```

```python
numerical_cols = ['SalePrice', 'GrLivArea', 'PoolArea']
```

```python
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], bins=30, kde=True, color='#1F4E79')
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
numerical_cols = ['SalePrice', 'GrLivArea', 'PoolArea']
```

```python
import numpy as np
# Buat DataFrame baru untuk data transformasi
df_log = df.copy()

for col in numerical_cols:
# Tambahkan 1 untuk menghindari log(0)
    df_log[col + '_log'] = np.log10(df_log[col] + 1)
```

```python
log_cols = [col + '_log' for col in numerical_cols]

for col in log_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df_log[col], bins=30, kde=True, color='#FFFF00')
    plt.title(f'Distribution of log10({col})')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
#Scatter plot GrLivArea vs SalePrice
plt.figure(figsize=(8,5))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df, color='#1F4E79')
plt.title('Relationship between GrLivArea and SalePrice')
plt.xlabel('GrLivArea')
plt.ylabel('SalePrice')
plt.show()
```

```python
#Scatter plot GarageArea vs SalePrice
plt.figure(figsize=(8,5))
sns.scatterplot(x='GarageArea', y='SalePrice', data=df, color='#FFFF00')
plt.title('Relationship between GarageArea and SalePrice')
plt.xlabel('GarageArea')
plt.ylabel('SalePrice')
plt.show()
```
